```python
import VariantCalling as vc
import matplotlib.pyplot as plt
```

```python
"""
This section to test ratio_gen, which is an alignment-level wrapper
"""
dg = vc.VariantCallingData()
align, prob = dg.ratio_gen(100,0.01,0.01)
align_int = dg.char_to_int(align)
plt.rcParams['figure.dpi'] = 200
#alignment_idx = mutation_types.index(mutation_index)
plt.title("Mutation")
plt.imshow(align_int,cmap='jet')
print(prob)
```
